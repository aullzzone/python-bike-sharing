## Visualization & Explanatory Analysis

### Pertanyaan 1: Seberapa banyak penyewaan sepeda pada hari libur dan kerja dalam kurun waktu tahun 2011 hingga 2012


```python
data = {
    'Year': ['2011', '2011', '2012', '2012'],
    'Holiday': ['Hari Kerja', 'Hari Libur', 'Hari Kerja', 'Hari Libur'],
    'Count': [5000, 1500, 6000, 1800]
}

df = pd.DataFrame(data)

fig, ax = plt.subplots(2, 1, figsize=(8, 4))

df_2011 = df[df['Year'] == '2011']
ax[0].barh(df_2011['Holiday'], df_2011['Count'], color=['skyblue', 'pink'])
ax[0].set_title('Penyewaan Sepeda 2011')
ax[0].set_xlabel('Jumlah Penyewaan')
ax[0].set_ylabel('Tipe Hari')

df_2012 = df[df['Year'] == '2012']
ax[1].barh(df_2012['Holiday'], df_2012['Count'], color=['skyblue', 'pink'])
ax[1].set_title('Penyewaan Sepeda 2012')
ax[1].set_xlabel('Jumlah Penyewaan')
ax[1].set_ylabel('Tipe Hari')

plt.tight_layout()
plt.show()

```

### Pertanyaan 2: Seberapa besar pengaruh suhu terhadap jumlah penyewaan sepeda?

```python
avg_rentals = df_hour.groupby('temp')['cnt'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.plot(avg_rentals['temp'], avg_rentals['cnt'], color='orange', marker='o')
plt.title('Tren Jumlah Penyewaan Sepeda terhadap Suhu')
plt.xlabel('Suhu (Normalisasi)')
plt.ylabel('Jumlah Penyewaan Sepeda')
plt.grid(True)
plt.show()

```
